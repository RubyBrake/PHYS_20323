1. Write a ``Particle`` class that can be used to represent a particle with a mass, a 3-d position, and a 3-d velocity.

5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np

class Particle(object):
    
    #all the variables and stuff
    def __init__(self, mass, px, py, pz, vx, vy, vz):
        self.mass = mass
        self.px = px
        self.py = py
        self.pz = pz
        self.vx = vx
        self.vy = vy
        self.vz = vz

    #calculating kinetic energy
    def kinetic(self):
         return 0.5*self.mass*((self.vx**2)+(self.vy**2)+(self.vz**2))

    #calculating distance
    def find_distance(self, other):
        return np.sqrt((self.px - other.px)**2 + (self.py - other.py)**2 + (self.pz - other.pz)**2)

    #updating position after a time interval
    def travel(self, dt):
        self.px += self.vx*dt
        self.py += self.vy*dt
        self.pz += self.vz*dt

# part 5
class ChargedParticle(Particle):
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        # Add a new attribute for electric charge
        self.charge = charge


#specific particles for use
p1 = Particle(10,0,0,0,1,1,1)
p2 = Particle(17,1,1,1,21,22,23)
p3 = ChargedParticle(17,1,1,1,21,22,23,-2)

#writing the answers
print("2. The kinetic energy with the random numbers I put in is", p1.kinetic())
print("3. The distance between the two particles I put in is", p1.find_distance(p3))
print ("4. The current position of particle 1 is", p1.px,p1.py,p1.pz)
p1.travel(3)
print ("4. The new position of particle 1 after 3 time units is", p1.px,p1.py,p1.pz)
print ("5. Particle 3 has position", p3.px, p3.py, p3.pz, "and charge", p3.charge)
```
